```python
from Sensei import *
```

```python
sensei = Sensei(42, stage=13) # things get hard around stage 13
```

```python
hc = sensei.practice()
```

```python
# There are two ways to probe the hidden circuit. You could add some operators at the beginning to see what it does.
pc = hc.get_empty_circuit()
hc.prepend(pc)
```

```python
# Another approach is to probe the hidden circuit by 'uncomputing' it by appending operators to the end. 
pc = hc.get_empty_circuit()
pc.t(0)
pc.h(0)
pc.x(0)
hc.append(pc)
```

```python
# Then, when you have gained some visual intuition, you can go for a final solution of replicating the circuit
sc = hc.get_empty_circuit()
hc.check(sc)
```

```python
sensei.submit(sc)
```
